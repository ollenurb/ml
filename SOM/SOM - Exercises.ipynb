## 1. Update rule for BMU(s)
Given the examples and the weights, show the updated weights of the BMUs

```python
import numpy as np

w = np.array([
    [0, 0],
    [0, 1],
    [1, 1.45],
    [3, 4]              
])

x = np.array([
    [1, 2],
    [1, 3],
    [-2, -3],
    [-5, -6],
    [5, 6],
    [4, 5]
])

eta = 0.5
for x_i in x:
  min_dist = np.linalg.norm(x_i - w, axis = 1)
  bmu_index = np.argmin(min_dist)
  # Apply the update rule for the BMU
  w[bmu_index] = w[bmu_index] + eta * (x_i - w[bmu_index]) 
  print(w[bmu_index])
```
